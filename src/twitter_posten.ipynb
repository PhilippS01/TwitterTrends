```python
web = 'https://twitter.com/login'

# Deine Twitter-Zugangsdaten
TWITTER_USERNAME = "@PhilippSch83993"
TWITTER_PASSWORD = "philippistreich"
```

```python
options = Options()
#options.add_argument("--headless")  # Ohne GUI starten
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# WebDriver starten
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

try:
    # Twitter Login-Seite öffnen
    driver.get(web)

    # Warte auf das Benutzername-Eingabefeld
    username_input = WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.XPATH, "//input[@type='text' and @name='text']")))
    username_input.send_keys(TWITTER_USERNAME)
    username_input.send_keys(Keys.RETURN)
    time.sleep(2)

    # Warte auf das Passwort-Eingabefeld
    password_input = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.NAME, "password"))
    )
    password_input.send_keys(TWITTER_PASSWORD)
    password_input.send_keys(Keys.RETURN)
    time.sleep(5)

    print("Erfolgreich eingeloggt!")

except Exception as e:
    import traceback
    print("Fehler aufgetreten:")
    print(traceback.format_exc())  # Gibt den gesamten Fehlerstack aus

finally:
    input("Drücke Enter zum Beenden...")
    # driver.quit()
```

### Analyse

```python
# Suchfeld finden
def suche_auf_X_nach(suche):
    search_input = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "input[data-testid='SearchBox_Search_Input']"))
    )
    print("Suchfeld gefunden!")

    # Suche starten
    search_input.send_keys(Keys.COMMAND + "a")
    #search_input.send_keys(Keys.CONTROL + "a")  # Windows/Linux
    search_input.send_keys(Keys.DELETE)
    search_input.send_keys(suche)
    search_input.send_keys(Keys.RETURN)
    print("Suche gestartet!")
```
